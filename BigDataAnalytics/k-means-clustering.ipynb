# K-Means Clustering on Iris Dataset
This notebook demonstrates clustering of Iris flowers using the K-Means algorithm. we use it without the labels to let k-means discover clusters on its own.



# What is K-Means?
K-Means groups data into k clusters based on similarity.

It doesn't use the target/label column, it figures out groups based on feature patterns.

Ideal for grouping unlabeled data.

```python
# import Libraries
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Load Iris data
iris = load_iris()
X = iris.data  # Only features
y = iris.target  # Labels for evaluation later
```

```python
# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
```

```python
# Predict clusters
clusters = kmeans.predict(X)
```

```python
# Reduce to 2D for visualization using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Plot the clusters
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=50)
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
